```python
import numpy as np
import pandas as pd
df=pd.read_excel(r"C:\Users\lenovo\OneDrive\Desktop\DataScience\OLA_trips_dataset.xlsx")
df.head(20)
```

```python
df.drop("booking id", axis=1, inplace=True)
df.drop("reason", axis=1, inplace=True)
df.drop("booking_date_time", axis=1, inplace=True)
df
```

```python
df.drop("gender", axis=1, inplace=True)
df.drop("commission_base_cost", axis=1, inplace=True)
df.drop("driver_base_cost", axis=1, inplace=True)
df
```

```python
df.drop("total_tax", axis=1, inplace=True)
df
```

```python
df.drop("time_of_day", axis=1, inplace=True)
```

```python
df.drop("ratings", axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['month']=lb.fit_transform(df['month'])
df['day_of_week']=lb.fit_transform(df['day_of_week'])
df['category']=lb.fit_transform(df['category'])
df.head(20)
```

```python
df.isnull().sum()
```

```python
y=df.iloc[:,6]
x=df.iloc[:,:6]
x
```

```python
import pickle
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model1 = LinearRegression(fit_intercept=True, n_jobs=None)
model1.fit(X_train, y_train)
print(X_train,y_train)
```

```python
pred1=model1.predict(X_test)
print(y_test,pred1)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
print("mean absolute error:",mean_absolute_error(y_test,pred1))
print("mean squared error:",mean_squared_error(y_test,pred1))
print("root mean squared error:",np.sqrt(mean_squared_error(y_test,pred1)))
print("r2-score:",r2_score(y_test,pred1))
```

```python
model1.predict([[5,5,17,58.00,0,2]])
```

```python
from sklearn.linear_model import Lasso
model = Lasso(alpha=0.0001)
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
print("mean absolute error:",mean_absolute_error(y_test,y_pred))
print("mean squared error:",mean_squared_error(y_test,y_pred))
print("root mean squared error:",np.sqrt(mean_squared_error(y_test,y_pred)))
print("r2-score:",r2_score(y_test,y_pred))
```

```python
model.predict([[5,5,17,58.00,0,2]])
```

```python
from xgboost import XGBRegressor
```

```python
xgb_model = XGBRegressor(random_state=42, n_estimators=100, learning_rate=0.05, max_depth=10,
                             min_child_weight=1, subsample=0.8, colsample_bytree=0.7)
xgb_model.fit(X_train, y_train)
y_pred6=xgb_model.predict(X_test)
```
